# stem plot

## syntax
import matplotlib.pyplot as plt

x=[ ]

y=[ ]

plt.stem(x,y)

plt.show()

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y,linefmt=':',markerfmt='r+')
plt.show()
```

## linefmt

- '-'    :- solid line
- '--'   :- Dashed line
- '-.'   :- Dash-dot line
- ':'    :- dotted line

```python
x=[1,2,3,4,5,6]
y=[2,2,5,6,4,3]
plt.stem(x,y,linefmt=':',markerfmt='ro',bottom=1,basefmt='g',label='python',
         orientation='horizontal')
plt.legend()
plt.show()
```

# Step Plot

## syntax
import matplotlib.pyplot as plt

x=[ ]

y=[ ]

plt.step(x,y)

plt.show()

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[3,2,3,5,4]

plt.step(x,y,color='r',marker='o',ms=10,mfc='b',label='python') # ms= marker size

plt.legend(loc=4)
plt.title('Python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid()
plt.show()
```

# Fill Between plot

## syntax
import matplotlib.pyplot as plt

x=[ ]

y=[ ]

plt.fill_between(x,y)

plt.show()



```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,2,3,4,5]

plt.plot(x,y,color='r')

plt.fill_between(x=[2,4],y1=2,y2=4)

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])

plt.plot(x,y,color='r')

plt.fill_between(x,y,color='red',
                 where=(x>=2)&  (x<=4),
                alpha=0.5)

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

# Savefig in Matplotlib Python

## import matplotlib.pyplot as plt

x=[ ] 

y=[ ] 

plt.plot(x,y)

plt.savefig(fname)

plt.show()


```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')
plt.savefig('line') # saves the graph
plt.show()
```

In matplotlib.pyplot.savefig(), the dpi parameter stands for Dots Per Inch.

Explanation:
dpi=2000 means you're saving the image at 2000 dots (pixels) per inch.

- It controls the resolution of the output image.
- Higher dpi means:
- More detailed and sharper image.
- Larger file size.
- Better quality for printing purposes.

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')
plt.savefig('line2',dpi=100,facecolor='g')  
plt.show()

```

```python
# saving in pdf format 
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[2,4,3,6,4]

plt.plot(x,y,color='r')
plt.savefig('line4',dpi=100,facecolor='g',
           transparent = True,bbox_inches='tight')  
plt.show()
```

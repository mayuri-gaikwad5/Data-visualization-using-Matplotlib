# Matplotlib Bar plot

## syntax
import matplotlib.pyplot as plt

x=[ ]

y=[ ]

plt.bar(x,y)

plt.show()

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]
y=[2,4,6,8]
c=['r','y','g','b']
plt.bar(x,y,color=c,alpha=0.4)# alpha value =0 to 1 to blur it

plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
plt.xlabel('language',fontsize=20)
plt.ylabel('Num.',fontsize=20)
plt.title('Bar Graph',fontsize=20)
c=['r','y','g','b']
plt.bar(x,y,width=0.4,color=c,align='edge',edgecolor='black',linewidth=2,linestyle='--'
       ,label='popularity')
plt.legend()
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]
plt.xlabel('language',fontsize=20)
plt.ylabel('Num.',fontsize=20)
plt.title('Bar Graph',fontsize=20)
plt.bar(x,y,width=0.4,color='r',label='popularity')
plt.bar(x,z,width=0.4,color='y',label='popularity1')
plt.legend()
plt.show()
```

```python
x=['python','c','c++','java']
import numpy as np

y=[85,70,60,82]
z=[20,30,40,50]
width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('language',fontsize=20)
plt.ylabel('Num.',fontsize=20)
plt.title('Bar Graph',fontsize=20)

plt.bar(p,y,width,color='r',label='popularity')
plt.bar(p1,z,width,color='y',label='popularity1')

plt.xticks(p+width/2,x,rotation=90)  #position,name,rotation
plt.legend()
plt.show()
```

```python
x=['python','c','c++','java']
y=[85,70,60,82]
z=[20,30,40,50]

plt.xlabel('language',fontsize=20)
plt.ylabel('Num.',fontsize=20)
plt.title('Bar Graph',fontsize=20)

plt.barh(x,y,color='r',label='popularity') # barh refers to horizontal
plt.barh(x,z,color='y',label='popularity1')

plt.legend()
plt.show()
```

# Scatter Plot

## syntax
import matplotlib.pyplot as plt

x=[ ]

y=[ ]

plt.scatter(x,y)

plt.show()

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[2,3,1,4,5,3,6]
colors=['r','y','g','b','r','g','r']
sizes=[100,600
       ,400,550,200,400,70]
plt.scatter(day,number,c=colors,s=sizes,alpha=0.4)
plt.title('Scatter Plot',fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('Number',fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[2,3,1,4,5,3,6]
colors=['r','y','g','b','r','g','r']
sizes=[100,600
       ,400,550,200,400,70]
plt.scatter(day,number,c=colors,s=sizes,marker='*',edgecolor='black',linewidth=2)
plt.title('Scatter Plot',fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('Number',fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[2,3,1,4,5,3,6]
colors=[10,40,30,29,56,20,30]
sizes=[100,600,400,550,200,400,70]

plt.scatter(day,number,c=colors,s=sizes,cmap='Oranges')
t=plt.colorbar()
t.set_label('color Bar',fontsize=15)
plt.title('Scatter Plot',fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('Number',fontsize=15)
plt.show()
```

```python
# multiple graphs
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
number=[2,3,1,4,5,3,6]
number2=[3,2,4,5,1,6,2]
colors=[10,40,30,29,56,20,30]
sizes=[100,600,400,550,200,400,70]

plt.scatter(day,number,c=colors,s=sizes)
plt.scatter(day,number2,c='r',s=sizes)
t=plt.colorbar()
t.set_label('color Bar',fontsize=15)
plt.title('Scatter Plot',fontsize=15)
plt.xlabel('Day',fontsize=15)
plt.ylabel('Number',fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,notch=True)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,vert=False,widths=0.4)

plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]

plt.boxplot(x,labels=['python'],patch_artist=True,
           showmeans=True)
# patch artist is used to fill the color
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70,120,200]

plt.boxplot(x,labels=['python'],showmeans=True,sym='g',
            boxprops=dict(color='r'),capprops=dict(color='purple'),
           whiskerprops=dict(color='b'))
# cap indicates maximum and minimun
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[10,20,30,40,50,60,70]
x1=[10,20,30,40,50,60,90]
y=[x,x1]

plt.boxplot(y,labels=['python','java'],showmeans=True,sym='g',
            boxprops=dict(color='r'),capprops=dict(color='purple'),
           whiskerprops=dict(color='b'))


plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
area=[2,3,2,5,4]
area1=[2,3,4,5,6]
area2=[1,3,2,4,2]

l=['area','area1','area2']
plt.stackplot(x,area,area1,area2,labels=l,colors=['r','g','m'],
              baseline='zero') # baseline can be;- zero,sym,wiggle

plt.title('python')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend(loc=2)

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ['c','c++','java','python']
exe = [0.4,0.0,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=exe,colors=c,autopct='%0.1f%%') 
# autopct for 

plt.show()
```

```python
#  adding shadow and adjusting the size
# setting the labels
# starting at different angles
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ['c','c++','java','python']
exe = [0.4,0.0,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=exe,colors=c,autopct='%0.1f%%',shadow=True,radius=0.8,
        labeldistance=1.1,startangle=90) 
# autopct for showing percentage

plt.show()
```

```python
# changing the font size of percentage using textprops
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ['c','c++','java','python']
exe = [0.4,0.0,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=exe,colors=c,autopct='%0.1f%%',shadow=True,
       textprops={'fontsize':15}) 
# you can even add font color

plt.show()
```

```python
# changing the rotation using counterclock 
# changing the width of the shadow using wedgeprops input should be given in the form of dictionary 
#  how to rotate labels
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ['c','c++','java','python']
exe = [0.4,0.0,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=exe,colors=c,autopct='%0.1f%%',shadow=True,
       textprops={'fontsize':15},counterclock=False,wedgeprops={'linewidth':5,'edgecolor':'black'}
       ,rotatelabels=True) 


plt.show()
```

```python
import matplotlib.pyplot as plt

x = [10,20,30,40]
y = ['c','c++','java','python']
exe = [0.4,0.0,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,explode=exe,colors=c,autopct='%0.1f%%',shadow=True,
       textprops={'fontsize':15},counterclock=False,
        wedgeprops={'linewidth':5,'edgecolor':'black'}) 

plt.title('Languages')
plt.legend(loc=2) # setting the location of labels
plt.show()
```

```python
# multiple pie charts
import matplotlib.pyplot as plt

x = [10,20,30,40]
x1=[40,30,20,10]
y = ['c','c++','java','python']

exe = [0.4,0.0,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,radius=1.5) 
plt.pie(x1,radius=0.5)


plt.show()
```

```python
# ring pie charts
import matplotlib.pyplot as plt

x = [10,20,30,40]
x1=[40,30,20,10]
y = ['c','c++','java','python']

exe = [0.4,0.0,0.0,0.0]
c=['r','b','g','y']

plt.pie(x,labels=y,radius=1.5) 
plt.pie([1],colors='w')


plt.show()
```
